```python
import re
```

```python
def tokenize(texto):
    tokens = []
    keywords = re.compile(r'\b(if|then|else|while)\b')
    identificador = re.compile(r'[a-zA-Z][a-zA-Z0-9]*')
    numeros = re.compile(r'\d+')
    operadores = re.compile(r'[\+\-\*\/]')
    delimitador = re.compile(r'[()]|;')
    # Recorrer palabra por palabra y asignar tipo
    for palabra in texto.split():
        if keywords.match(palabra):
            tokens.append(('Keyword', palabra))
        elif identificador.match(palabra):
            tokens.append(('Identificador', palabra))
        elif numeros.match(palabra):
            tokens.append(('Número', palabra))
        elif operadores.match(palabra):
            tokens.append(('Operador', palabra))
        elif delimitador.match(palabra):
            tokens.append(('Delimitador', palabra))
        else:
            print(f'Token desconocido: {palabra}')
    return tokens
            
```

```python
# Ejemplo de uso
texto = 'if x > 5 then ( y = y + 1 ) ; else z = z - 1 ; @'
tokens = tokenize(texto)
print(tokens)
```
