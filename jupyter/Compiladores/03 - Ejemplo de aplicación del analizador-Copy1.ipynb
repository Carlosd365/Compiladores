```python
import lexico
```

```python
# Ejemplo de uso
codigo_fuente = """
int suma(int a, int b){
    return a - b;
}
"""
```

```python
# Analisis lexico
tokens = lexico.identificar_tokens(codigo_fuente)
tokens
```

```python
print('Tokens encontrados')
for tipo, valor in tokens:
    print(f'{tipo}: {valor}')
```

```python
import sintactico
```

```python
# Analisis sintactico
try:
    print('\nIniciando analisis sintactico...')
    parser = sintactico.Parser(tokens)
    parser.parsear()
    print('Analisis sintactico completo sin errores!')
except SyntaxError as e:
    print(e)
```
