```python
import pandas as pd
import re
from PyPDF2 import PdfReader

rutaArch = "Analisis Anual 2018 ETAS.pdf"

documento = PdfReader(rutaArch)

dir(documento)

len(documento.pages)

pagina = documento.pages[4]
texto = pagina.extract_text() 
texto

patron = re.compile(r'\n(\d+)\s+(\d+)\s+(\w+\s*\w*\W*\s*\w*\s*\w*)\s+(\w+\s*\w*)\s+(\W*\w*\W*\w*\s*\w*)\s+(\d*)')

datos = re.findall(patron, texto)
datos

etas = pd.DataFrame(datos, columns=['No', 'Se', 'Área de Salud', 'ETA reportada ', 'Fuente de Contagio', 'No. De casos '])
etas
```
